# Training dataset

```python
save_dir = "/data/processed/bboxes_available/png_slices/training_set_images/"
```

```python
df = read_boxes(boxes_fp="/data/raw/BCS-DBT boxes-train-v2.csv", filepaths_fp="/data/raw/BCS-DBT file-paths-train-v2.csv")
```

```python
df.head()
```

```python
df.Class.value_counts()
```

For all DICOM volumes that have bounding boxes available, save each slice as an individual png image.

```python
for i in range(df.shape[0]):
    box_series = df.iloc[i]
    view = box_series["View"]
    image_path = os.path.join("/data/raw/training_set_images/", box_series["descriptive_path"])
    image = dcmread_image(fp=image_path, view=view)
    n_slices = image.shape[0]
    for j in range(n_slices):
        file_name = box_series["descriptive_path"][28:-4].replace("/", "__")
        file_name = os.path.join(save_dir, "{}__slice-{}.png".format(file_name, j))
        Image.fromarray(image[j]).save(file_name)
        print("{}/{}: slice {}/{} saved as png.".format(i, df.shape[0], j, n_slices))
```
